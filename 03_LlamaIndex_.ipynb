# `What is LLAMA-INDEX:`
LlamaIndex, formerly known as GPT Index, is a tool that helps you connect large language models (like GPT) with your own data. It creates a special index, kind of like a detailed table of contents, that helps the model quickly find and use the information it needs from your data. This makes it easier to ask questions or get insights from your data using AI


`LlamaIndex `is about connecting LLMs with your data by creating an index.

`LangChain` is about building complex applications with LLMs by managing sequences of operations or interactions.
